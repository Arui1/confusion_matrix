Demo of make_confusion_matrix function.

```python
import numpy as np
from cf_matrix import make_confusion_matrix
```

```python
cf_matrix = np.array([[23,  5],
                      [ 3, 30]])
```

```python
labels = ['True Neg','False Pos','False Neg','True Pos']
categories = ['Zero', 'One']
make_confusion_matrix(cf_matrix, group_names=labels, categories=categories, cmap='binary')
```

Multi-Categorical Confusion Matrix

```python
cf_matrix_3x3 = np.array([[ 856,  58,   130],
                          [  35,  765,  136],
                          [  69,  33, 907]])
```

```python
make_confusion_matrix(cf_matrix_3x3)
```
